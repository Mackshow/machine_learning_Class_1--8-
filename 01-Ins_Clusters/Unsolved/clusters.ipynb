```python
# Import the modules
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate three synthetic clusters
X, y = make_blobs(n_samples=300,
    centers=3, 
    random_state=0, 
    cluster_std=0.60
)
```

```python
# Visualizing all three classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```
